### THEORY QUESTIONS:

1. What is hypothesis testing in statistics ?
   - Hypothesis testing in statistics is a method used to make decisions or inferences about a population based on sample data.
**In short**:
   - You start with a null hypothesis (H₀) — a statement of no effect or no difference.

   - You also have an alternative hypothesis (H₁) — what you want to test for.

   - You collect sample data and calculate a test statistic.

   - Based on this, you calculate a p-value or compare with a critical value.

   - If the evidence is strong enough (usually if p-value < significance level, e.g., 0.05), you reject H₀; otherwise, you fail to reject H₀.

2. What is the null hypothesis, and how does it differ from the alternative hypothesis ?
   - Null hypothesis (H₀): Assumes no effect or no difference.

   - Alternative hypothesis (H₁): Assumes there is an effect or a difference.

   - We test data to see if there’s enough evidence to reject H₀ in favor of H₁.

3. What is the significance level in hypothesis testing, and why is it important ?
    - The significance level (α) is the threshold for deciding whether to reject the null hypothesis.

    - Commonly set at 0.05 (5%), meaning a 5% risk of rejecting H₀ when it’s actually true (Type I error).

    - Important because it controls the probability of making false positive conclusions.

4. What does a P-value represent in hypothesis testing ?
   - The P-value is the probability of getting the observed data (or more extreme) if the null hypothesis (H₀) is true.

   - A small P-value (e.g., < 0.05) suggests strong evidence against H₀.

   - A large P-value suggests insufficient evidence to reject H₀.

5.  How do you interpret the P-value in hypothesis testing ?
   - If P-value ≤ α (e.g., 0.05): Reject H₀ → evidence supports the alternative hypothesis.

   - If P-value > α: Fail to reject H₀ → not enough evidence to support the alternative hypothesis.

6. What are Type 1 and Type 2 errors in hypothesis testing ?
   - Type I Error (False Positive): Rejecting H₀ when it is actually true.

   - Type II Error (False Negative): Failing to reject H₀ when it is actually false.

   - Type I error risk = significance level (α).
   - Type II error risk = denoted by β.

7. What is the difference between a one-tailed and a two-tailed test in hypothesis testing ?
   - One-tailed test: Tests for an effect in one specific direction (e.g., greater than or less than).

   - Two-tailed test: Tests for an effect in both directions (e.g., not equal to).

8. What is the Z-test, and when is it used in hypothesis testing ?
    - The Z-test is a statistical test used to determine if there’s a significant difference between sample and population means (or between two sample means) when:

         - The population variance is known, and

         - The sample size is large (typically n ≥ 30).

    - It uses the Z-score, which measures how many standard deviations a data point is from the population mean.



9. How do you calculate the Z-score, and what does it represent in hypothesis testing ?
    -  **Z-score formula:**

$$
Z = \frac{\bar{X} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$
  
  where:
   - $\bar{X}$ = sample mean
   - $\mu$ = population mean
   - $\sigma$ = population standard deviation
   - $n$ = sample size

  -  **What it represents:**
  The **Z-score** tells how many standard deviations the sample mean is away from the population mean.

 - In hypothesis testing, it helps determine how unusual the sample result is under the null hypothesis.


10. What is the T-distribution, and when should it be used instead of the normal distribution ?
   - The T-distribution is similar to the normal distribution but has heavier tails (more variability).

   - Use it when:

      - The sample size is small (typically n < 30)

      - The population standard deviation is unknown

    - As sample size increases, the T-distribution approaches the normal distribution.

11. What is the difference between a Z-test and a T-test ?
    - Z-test:

       - Used when population standard deviation (σ) is known

       - Usually for large samples (n ≥ 30)

       - Based on normal distribution

    - T-test:

        - Used when population standard deviation is unknown

        - Usually for small samples (n < 30)

        - Based on T-distribution with heavier tails

12. What is the T-test, and how is it used in hypothesis testing ?
   - The T-test compares sample data to a population or between two samples when the population standard deviation is unknown.

   - It calculates a T-score to see if differences are statistically significant.

   - Common types:

       - One-sample T-test: compare sample mean to population mean.

       - Independent T-test: compare means of two independent groups.

       - Paired T-test: compare means from the same group at different times.

    - Used in hypothesis testing to decide whether to reject H₀.

13. What is the relationship between Z-test and T-test in hypothesis testing ?
   - Both Z-test and T-test compare means to test hypotheses.

   - They differ mainly in sample size and knowledge of population standard deviation:

       - Use Z-test when σ (population standard deviation) is known and sample size is large.

       - Use T-test when σ is unknown and/or sample size is small.

    - As sample size grows, the T-distribution approaches the normal distribution, making Z-test and T-test results very similar for large samples.

14. What is a confidence interval, and how is it used to interpret statistical results ?
 - A confidence interval (CI) gives a range of values that likely contains the true population parameter.

     - For example, a 95% CI means we are 95% confident the true value lies within that range.

     - Used to interpret results by showing estimate precision:

          - Narrow CI: more precise estimate.

          - Wide CI: less precise estimate.

     - If the CI does not include the null value (e.g., 0 for differences), it suggests a significant result.










15. What is the margin of error, and how does it affect the confidence interval ?
- The **margin of error (MOE)** is the amount added and subtracted from the sample estimate to create the confidence interval.
- Formula:

$$
\text{MOE} = \text{critical value} \times \text{standard error}
$$

- **Affects the confidence interval** by determining its width:

  - **Larger MOE → wider CI → less precision**
  - **Smaller MOE → narrower CI → more precision**
- MOE depends on sample size, variability, and confidence level.









16. How is Bayes' Theorem used in statistics, and what is its significance ?
  
- **Bayes' Theorem** updates the probability of an event based on new evidence.
- Formula:

$$
P(A|B) = \frac{P(B|A) \cdot P(A)}{P(B)}
$$

- **In statistics:**

  - Used in **Bayesian inference** to revise beliefs (probabilities) after seeing data.
  - Helps in fields like medical testing, machine learning, and decision-making.

**Significance:**

- Allows combining prior knowledge with new data.
- Gives a flexible way to handle uncertainty and update probabilities.


17. What is the Chi-square distribution, and when is it used ?

- The **Chi-square distribution** is a distribution of the sum of squared standard normal variables.
- **Used when:**

  - Testing relationships between categorical variables (**Chi-square test of independence**)
  - Testing goodness of fit (**Chi-square goodness-of-fit test**)
  - Analyzing variance in sample data (**variance tests**)
- It’s **always positive** and **skewed**, especially for small degrees of freedom.


18. What is the Chi-square goodness of fit test, and how is it applied ?

- The **Chi-square goodness of fit test** checks if observed data matches an expected distribution.
- **Steps:**

  1. Set null hypothesis (H₀): observed data fits expected distribution.
  2. Calculate test statistic:

$$
\chi^2 = \sum \frac{(O - E)^2}{E}
$$

where $O$ = observed frequency, $E$ = expected frequency.
  3. Compare with critical value or compute p-value.

- **Applied in:** checking if data fits a theoretical model (e.g., dice fairness, genetic ratios).


19. What is the F-distribution, and when is it used in hypothesis testing ?

- The **F-distribution** is a ratio of two variances (each scaled by their degrees of freedom).
- **Used in hypothesis testing for:**

  - **ANOVA (Analysis of Variance):** to compare means of multiple groups.
  -  **Variance comparison tests:** to test if two populations have equal variances.
-  It is **skewed right** and depends on two sets of degrees of freedom.


20. What is an ANOVA test, and what are its assumptions ?

- **ANOVA (Analysis of Variance):** tests whether **three or more group means** are significantly different.
- It uses the **F-distribution** to compare between-group and within-group variance.

**Assumptions:**

1. **Independence** of observations.
2. **Normality** of data within groups.
3. **Equal variances** (homogeneity of variance) across groups.


21. What are the different types of ANOVA tests ?

  1. **One-Way ANOVA:**

     -  Tests differences between means of **one factor** with multiple groups.

  2. **Two-Way ANOVA:**

     - Tests effects of **two factors** and their interaction on the response variable.

  3. **Repeated Measures ANOVA:**

     - Used when the **same subjects** are measured multiple times under different conditions.

  4. **MANOVA (Multivariate ANOVA):**

     - Extends ANOVA to test multiple dependent variables simultaneously.


22. What is the F-test, and how does it relate to hypothesis testing ?

- The **F-test** compares **two variances** to see if they are significantly different.
- Formula:

$$
F = \frac{\text{Variance}_1}{\text{Variance}_2}
$$

- **In hypothesis testing:**

  - Used in **ANOVA** to compare group variances.
  - Tests the null hypothesis that variances (or means in ANOVA) are equal.
- A large F-value suggests evidence to **reject H₀**.

2. Generate a discrete uniform distribution using Python and plot the probability mass function (PMF) .

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

low = 1
high = 6
size = 1000

samples = np.random.randint(low, high + 1, size)

counts = Counter(samples)
values = sorted(counts.keys())
frequencies = [counts[v]/size for v in values]  # Relative frequency as PMF

plt.stem(values, frequencies, basefmt=" ")
plt.title("PMF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Probability")
plt.xticks(values)
plt.grid(True)
plt.show()
```

3.  Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution .

```python
import matplotlib.pyplot as plt

def bernoulli_pmf(p):
    """
    p: probability of success (0 <= p <= 1)
    """
    x = [0, 1]
    pmf = [1 - p, p]

    for xi, pi in zip(x, pmf):
        print(f"P(X = {xi}) = {pi}")

    plt.stem(x, pmf, basefmt=" ")
    plt.title(f"Bernoulli Distribution (p={p})")
    plt.xlabel("Outcome")
    plt.ylabel("Probability")
    plt.xticks(x)
    plt.ylim(0, 1.1)
    plt.grid(True)
    plt.show()

bernoulli_pmf(0.7)
```

6. Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete
uniform distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

low = 1
high = 6
x = np.arange(low, high + 1)
pmf = np.full_like(x, 1/(high - low + 1), dtype=float)

cdf = np.cumsum(pmf)

plt.step(x, cdf, where='post', color='blue', label='CDF')
plt.title("CDF of Discrete Uniform Distribution")
plt.xlabel("Value")
plt.ylabel("Cumulative Probability")
plt.xticks(x)
plt.ylim(0, 1.1)
plt.grid(True)
plt.legend()
plt.show()
```

11. Simulate multiple samples from a normal distribution and verify the Central Limit Theorem .

14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal
distribution .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def calculate_z_score(data_point, mean, std_dev):
    z = (data_point - mean) / std_dev
    return z
data_point = 75
mean = 70
std_dev = 10

z_score = calculate_z_score(data_point, mean, std_dev)
print(f"Data Point: {data_point}")
print(f"Mean: {mean}, Standard Deviation: {std_dev}")
print(f"Z-score: {z_score:.2f}")

x = np.linspace(-4, 4, 1000)
pdf = norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, pdf, label='Standard Normal PDF', color='blue')
plt.fill_between(x, pdf, where=(x <= z_score), color='orange', alpha=0.5, label='Area ≤ Z')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z = {z_score:.2f}')
plt.title("Standard Normal Distribution")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

15. Implement hypothesis testing using Z-statistics for a sample dataset .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
population_mean = 100
population_std = 15
sample_size = 50
np.random.seed(0)
sample = np.random.normal(population_mean, population_std, sample_size)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
standard_error = population_std / np.sqrt(sample_size)
z_statistic = (sample_mean - population_mean) / standard_error

p_value = 2 * (1 - norm.cdf(abs(z_statistic)))

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_statistic:.2f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Reject the null hypothesis (significant difference).")
else:
    print("Fail to reject the null hypothesis (no significant difference).")
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(8,5))
plt.plot(x, y, label='Standard Normal Distribution')

plt.fill_between(x, y, where=(x <= norm.ppf(alpha/2)), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x >= norm.ppf(1 - alpha/2)), color='red', alpha=0.3)

plt.axvline(z_statistic, color='blue', linestyle='--', label=f'Z = {z_statistic:.2f}')
plt.title("Z-Test Hypothesis Testing")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

16. Create a confidence interval for a dataset using Python and interpret the result .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(0)
sample_size = 50
population_mean = 100
population_std = 15
sample = np.random.normal(population_mean, population_std, sample_size)
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)

confidence_level = 0.95
alpha = 1 - confidence_level

standard_error = sample_std / np.sqrt(sample_size)
z_critical = norm.ppf(1 - alpha/2)
margin_of_error = z_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Confidence Interval ({int(confidence_level*100)}%): ({lower_bound:.2f}, {upper_bound:.2f})")

print("\nInterpretation:")
print(f"We are {int(confidence_level*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")

x = np.linspace(sample_mean - 4*standard_error, sample_mean + 4*standard_error, 100)
y = norm.pdf(x, sample_mean, standard_error)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Sampling Distribution')
plt.axvline(lower_bound, color='red', linestyle='--', label='Confidence Interval')
plt.axvline(upper_bound, color='red', linestyle='--')
plt.axvline(sample_mean, color='blue', linestyle='-', label='Sample Mean')
plt.fill_between(x, y, where=(x >= lower_bound) & (x <= upper_bound), color='orange', alpha=0.3)
plt.title(f"{int(confidence_level*100)}% Confidence Interval for Mean")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

np.random.seed(0)
sample_size = 30
population_mean = 100
population_std = 15
sample = np.random.normal(population_mean, population_std, sample_size)

sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)
confidence_level = 0.95
alpha = 1 - confidence_level
degrees_of_freedom = sample_size - 1

standard_error = sample_std / np.sqrt(sample_size)
t_critical = stats.t.ppf(1 - alpha/2, df=degrees_of_freedom)
margin_of_error = t_critical * standard_error

lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Sample Standard Deviation: {sample_std:.2f}")
print(f"Confidence Interval ({int(confidence_level*100)}%): ({lower_bound:.2f}, {upper_bound:.2f})")
print("\nInterpretation:")
print(f"We are {int(confidence_level*100)}% confident that the true population mean lies between {lower_bound:.2f} and {upper_bound:.2f}.")
x = np.linspace(sample_mean - 4*standard_error, sample_mean + 4*standard_error, 100)
y = stats.t.pdf((x - sample_mean) / standard_error, df=degrees_of_freedom)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='t-distribution (sampling distribution)')
plt.axvline(lower_bound, color='red', linestyle='--', label='Confidence Interval')
plt.axvline(upper_bound, color='red', linestyle='--')
plt.axvline(sample_mean, color='blue', linestyle='-', label='Sample Mean')
plt.fill_between(x, y, where=(x >= lower_bound) & (x <= upper_bound), color='orange', alpha=0.3)
plt.title(f"{int(confidence_level*100)}% Confidence Interval for Mean (t-distribution)")
plt.xlabel("Value")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

1. Write a Python program to perform a Z-test for comparing a sample mean to a known population mean and interpret the results .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
np.random.seed(0)
sample_size = 50
sample_mean_true = 102
population_std = 10
sample = np.random.normal(sample_mean_true, population_std, sample_size)

population_mean = 100
sample_mean = np.mean(sample)
standard_error = population_std / np.sqrt(sample_size)
z_stat = (sample_mean - population_mean) / standard_error
p_value = 2 * (1 - norm.cdf(abs(z_stat)))
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (significant difference).")
else:
    print("Result: Fail to reject the null hypothesis (no significant difference).")
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x, 0, 1)

plt.figure(figsize=(8,5))
plt.plot(x, y, label='Standard Normal Distribution')
plt.fill_between(x, y, where=(x <= norm.ppf(alpha/2)), color='red', alpha=0.3, label='Rejection Region')
plt.fill_between(x, y, where=(x >= norm.ppf(1 - alpha/2)), color='red', alpha=0.3)
plt.axvline(z_stat, color='blue', linestyle='--', label=f'Z = {z_stat:.2f}')
plt.title("One-sample Z-test")
plt.xlabel("Z-score")
plt.ylabel("Density")
plt.legend()
plt.grid(True)
plt.show()
```

5.  Create a Python function that calculates and visualizes Type 1 and Type 2 errors during hypothesis testing .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

def visualize_type1_type2(mu_null, mu_alt, sigma, n, alpha):

    se = sigma / np.sqrt(n)

    z_critical = norm.ppf(1 - alpha / 2)

    lower_critical = mu_null - z_critical * se
    upper_critical = mu_null + z_critical * se

    beta_lower = norm.cdf(lower_critical, mu_alt, se)
    beta_upper = 1 - norm.cdf(upper_critical, mu_alt, se)
    beta = beta_lower + beta_upper

    print(f"Type I Error (alpha): {alpha:.2f}")
    print(f"Type II Error (beta): {beta:.4f}")

    x = np.linspace(mu_null - 4*se, mu_alt + 4*se, 1000)

    plt.figure(figsize=(10,6))

    plt.plot(x, norm.pdf(x, mu_null, se), label="H₀: Null Distribution", color='blue')

    plt.plot(x, norm.pdf(x, mu_alt, se), label="H₁: Alternative Distribution", color='orange')

    plt.fill_between(x, 0, norm.pdf(x, mu_null, se), where=(x <= lower_critical), color='red', alpha=0.3, label='Type I Error (α)')
    plt.fill_between(x, 0, norm.pdf(x, mu_null, se), where=(x >= upper_critical), color='red', alpha=0.3)

    plt.fill_between(x, 0, norm.pdf(x, mu_alt, se), where=(x >= lower_critical) & (x <= upper_critical), color='green', alpha=0.3, label='Type II Error (β)')

    plt.axvline(lower_critical, color='black', linestyle='--', label='Critical Values')
    plt.axvline(upper_critical, color='black', linestyle='--')

    plt.title("Type I and Type II Errors Visualization")
    plt.xlabel("Sample Mean")
    plt.ylabel("Density")
    plt.legend()
    plt.grid(True)
    plt.show()

mu_null = 100
mu_alt = 105
sigma = 10
n = 30
alpha = 0.05

visualize_type1_type2(mu_null, mu_alt, sigma, n, alpha)
```

9.  Write a Python function to calculate the confidence interval for a sample mean and explain its significance .

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(sample, confidence=0.95):
    n = len(sample)
    mean = np.mean(sample)
    std_err = stats.sem(sample)
    h = std_err * stats.t.ppf((1 + confidence) / 2, n - 1)
    lower = mean - h
    upper = mean + h
    return mean, lower, upper

np.random.seed(0)
sample_size = 30
true_mean = 100
true_std = 15

sample = np.random.normal(true_mean, true_std, sample_size)

mean, lower, upper = confidence_interval(sample, confidence=0.95)

print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: [{lower:.2f}, {upper:.2f}]")

plt.figure(figsize=(8, 5))
plt.errorbar(1, mean, yerr=(upper - mean), fmt='o', color='blue', capsize=5, label='95% CI')
plt.scatter(1, mean, color='red', label=f"Mean = {mean:.2f}")
plt.axhline(true_mean, color='green', linestyle='--', label="True Mean")
plt.xlim(0, 2)
plt.title("Confidence Interval for Sample Mean")
plt.ylabel("Value")
plt.xticks([])
plt.legend()
plt.grid(True)
plt.show()
```

10. Write a Python program to calculate the margin of error for a given confidence level using sample data .

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def calculate_margin_of_error(sample, confidence=0.95):
    n = len(sample)
    std_err = stats.sem(sample)
    t_critical = stats.t.ppf((1 + confidence) / 2, n - 1)
    margin_of_error = t_critical * std_err
    return margin_of_error

np.random.seed(0)
sample_size = 40
true_mean = 100
true_std = 15
sample = np.random.normal(true_mean, true_std, sample_size)

confidence_level = 0.95
margin_of_error = calculate_margin_of_error(sample, confidence_level)

sample_mean = np.mean(sample)
lower_bound = sample_mean - margin_of_error
upper_bound = sample_mean + margin_of_error

print(f"Sample Mean: {sample_mean:.2f}")
print(f"Margin of Error (95% CI): ±{margin_of_error:.2f}")
print(f"95% Confidence Interval: [{lower_bound:.2f}, {upper_bound:.2f}]")

plt.figure(figsize=(8, 5))
plt.errorbar(1, sample_mean, yerr=margin_of_error, fmt='o', color='blue', capsize=5, label='95% CI')
plt.scatter(1, sample_mean, color='red', label=f"Mean = {sample_mean:.2f}")
plt.axhline(true_mean, color='green', linestyle='--', label="True Mean")
plt.xlim(0, 2)
plt.title("Margin of Error Visualization")
plt.ylabel("Value")
plt.xticks([])
plt.legend()
plt.grid(True)
plt.show()
```

11.  Implement a Bayesian inference method using Bayes' Theorem in Python and explain the process .

```python
import numpy as np
import matplotlib.pyplot as plt

P_disease = 0.01
P_no_disease = 1 - P_disease

sensitivity = 0.99
specificity = 0.95
false_positive_rate = 1 - specificity
numerator = sensitivity * P_disease

denominator = (sensitivity * P_disease) + (false_positive_rate * P_no_disease)

P_disease_given_positive = numerator / denominator

print(f"Prior probability of disease: {P_disease:.2%}")
print(f"Posterior probability of disease given positive test: {P_disease_given_positive:.2%}")

labels = ['Disease', 'No Disease']
priors = [P_disease, P_no_disease]
posteriors = [P_disease_given_positive, 1 - P_disease_given_positive]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, priors, width, label='Prior')
plt.bar(x + width/2, posteriors, width, label='Posterior')
plt.xticks(x, labels)
plt.ylabel('Probability')
plt.title("Bayesian Inference: Prior vs Posterior")
plt.legend()
plt.grid(True, axis='y')
plt.show()
```

12. Perform a Chi-square test for independence between two categorical variables in Python .

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import chi2_contingency

np.random.seed(0)
n_samples = 200
gender = np.random.choice(['Male', 'Female'], size=n_samples, p=[0.5, 0.5])
preference = np.random.choice(['A', 'B'], size=n_samples, p=[0.6, 0.4])

data = pd.DataFrame({'Gender': gender, 'Preference': preference})
contingency_table = pd.crosstab(data['Gender'], data['Preference'])
print("Contingency Table:")
print(contingency_table)

chi2, p, dof, expected = chi2_contingency(contingency_table)

print("\nChi-square Statistic:", chi2)
print("Degrees of Freedom:", dof)
print("P-value:", p)
print("\nExpected Frequencies:")
print(pd.DataFrame(expected, index=contingency_table.index, columns=contingency_table.columns))

alpha = 0.05
if p < alpha:
    print("\nResult: Reject the null hypothesis (variables are dependent).")
else:
    print("\nResult: Fail to reject the null hypothesis (variables are independent).")

contingency_table.plot(kind='bar', stacked=True)
plt.title("Contingency Table Visualization")
plt.ylabel("Count")
plt.grid(axis='y')
plt.show()
```

14. Perform a goodness-of-fit test using Python to compare the observed data to an expected distribution .

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import chisquare

observed = np.array([18, 22, 20, 21, 19, 20])

expected = np.full_like(observed, fill_value=np.sum(observed) / len(observed))

chi2_stat, p_value = chisquare(f_obs=observed, f_exp=expected)
print("Observed Frequencies:", observed)
print("Expected Frequencies:", expected)
print(f"Chi-square Statistic: {chi2_stat:.4f}")
print(f"P-value: {p_value:.4f}")
alpha = 0.05
if p_value < alpha:
    print("Result: Reject the null hypothesis (distribution is not a good fit).")
else:
    print("Result: Fail to reject the null hypothesis (distribution fits well).")

faces = ['1', '2', '3', '4', '5', '6']
x = np.arange(len(faces))
width = 0.35

plt.figure(figsize=(8,5))
plt.bar(x - width/2, observed, width, label='Observed', color='skyblue', edgecolor='black')
plt.bar(x + width/2, expected, width, label='Expected', color='orange', edgecolor='black')
plt.xticks(x, faces)
plt.title("Chi-square Goodness-of-Fit: Die Rolls")
plt.xlabel("Die Face")
plt.ylabel("Frequency")
plt.legend()
plt.grid(axis='y')
plt.show()
```
